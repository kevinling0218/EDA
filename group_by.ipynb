```python
import os 
import pandas as pd
import numpy as np
```

```python
DATA_FOLDER = './housing_price/data'
df = pd.read_csv(os.path.join(DATA_FOLDER, 'train.csv'))
```

```python
df.head()
```

```python
# First entry of each year
df.groupby('YrSold').first()
```

```python
# Not using groupby object as index 
df.groupby('YrSold',as_index = False).first()
```

```python
# Sum of the sales per year, same for count(),max(),min()
df[df['SaleCondition']=='Normal'].groupby('YrSold')['SalePrice'].sum()
```

```python
# Using mean and len to get a count 
pd.pivot_table (df,index = ['YrSold','SaleCondition'],values = ['SalePrice'],aggfunc=[np.median,len])
```

```python
# Using groupby is somehow similar with the pivot_table result
df.groupby(['YrSold','SaleCondition'])['SalePrice'].median()
```

```python
# Using different method for different feature
df.groupby('YrSold').agg({'SalePrice':'sum','LotArea':'mean'})
```

```python
# multiple statistical method
# Lambda function to extract statistics that are not provided by the built-in
aggregation = {
    'SalePrice':[min,max,'mean'],
    'MoSold':'count',
    'LotArea':lambda x: max(x)/1000
}
df.groupby('YrSold').agg(aggregation)
```

```python
# Rename the result
grouped = df.groupby('YrSold').agg({'SalePrice':[min,max,'mean']})
grouped.rename(columns={"min":"MinPrice","max":"MaxPrice","mean":"AvergaePrice"})
```

```python
# Using ravel
grouped = df.groupby('YrSold').agg({'SalePrice':[min,max,'mean']})
grouped.columns = ["_".join(x) for x in grouped.columns.ravel()]
grouped
```
