```python
### Import modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
# Comment this if the data visualisations doesn't work on your side
%matplotlib inline
```

```python
### Read data

DATA_FOLDER = './data/'

df_train = pd.read_csv(os.path.join(DATA_FOLDER, 'train.csv'))
```

# Get basic info for data

```python
df_train.info()
```

```python
df_train.isna().sum()
```

From the data we know there are quite a number of columns has too many null values, so drop the column which has more than 30% of NaN value

```python
# Remove if the NaN value is more than 30%
df_train_2 = df_train[[column for column in df_train 
                       if df_train[column].count()/len(df_train)>=0.3]]
df_train_2.drop(['Id'],axis = 1)
```

# Statistics for target column

```python
print (df_train_2.SalePrice.describe())
plt.figure(figsize= (9,8))
sns.distplot(df_train_2.SalePrice, color = 'g', bins = 20,hist_kws={'alpha': 0.4})
```

# Numerical Data Distribution

```python
### Select only numeric data
df_num = df_train_2.select_dtypes(include = ['float64','int64'])
```

```python
### Plot the distribution of all features

df_num.hist(figsize = (16,20), bins = 20, xlabelsize=8, ylabelsize=8)
```

```python
### Correlation

# get the correlation between SalePrice with other columns
df_num_corr = df_num[df_num.columns[1:-1]].apply(lambda x: x.corr(df_num['SalePrice'])) 
```

```python
df_num_corr
```

```python
# Get strong correlation features
strong_corr_features = df_num_corr[abs(df_num_corr)>0.5].sort_values(ascending=False)

print('Features: \n{}'.format(strong_corr_features))
```

```python
# Plot the correlation between features and target 
# Since the linear correlation number may be affected by outliers
for i in range (0, len(df_num.columns), 5):
    sns.pairplot(data = df_num,
                x_vars = df_num.columns[i:i+5],
                y_vars = ['SalePrice'])
```

```python
# Remove 0 values and re-calculate the correaltion
# Since the correlation is highly affected by 0 values

import operator
individual_features_df = []
for i in range (0,len(df_num.columns)-1):
    tmpDf = df_num[[df_num.columns[i],'SalePrice']]
    tmpDf = tmpDf[tmpDf[df_num.columns[i]]!= 0]
    individual_features_df.append(tmpDf)
    
all_corr = {feature.columns[0]:feature.corr()['SalePrice'][0] for feature in 
           individual_features_df}
all_corr = sorted(all_corr.items(), key=operator.itemgetter(1))
for (key,value) in all_corr:
    print ("{:>15}: {:>15}".format(key, value))
```

```python
# Get another set of golden features which has corr >0.5 excluding 0
golden_features_list = [key for key, value in 
                        all_corr if abs(value) >= 0.5]
print("There is {} strongly correlated values with SalePrice:\n{}"
      .format(len(golden_features_list), golden_features_list))
```
