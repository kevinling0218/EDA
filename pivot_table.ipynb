```python
import os 
import pandas as pd
import numpy as np
```

```python
DATA_FOLDER = './housing_price/data'
df = pd.read_csv(os.path.join(DATA_FOLDER, 'train.csv'))
```

```python
df.head()
```

```python
pd.pivot_table (df,index = ['YrSold','SaleCondition'])
```

```python
# Default aggregation method is mean value 
pd.pivot_table (df,index = ['YrSold','SaleCondition'],values = ['SalePrice'])
```

```python
# Using mean and len to get a count 
pd.pivot_table (df,index = ['YrSold','SaleCondition'],values = ['SalePrice'],aggfunc=[np.median,len])
```

```python
# Further segragating using column 
# fill_value to deal with NaN problem
pd.pivot_table (df,index = ['YrSold','SaleCondition'],
                columns = ['Utilities'],values = ['SalePrice'],
                aggfunc=[np.median,len],fill_value = 0)
```

```python
# The result is still a DataFrame

table = pd.pivot_table (df,index = ['YrSold','SaleCondition'],
                columns = ['Utilities'],values = ['SalePrice'],
                aggfunc=[np.median,len],fill_value = 0)
```

```python
# Only look at certain SaleCondition
table.query('SaleCondition == ["Normal","Family"]')
```
